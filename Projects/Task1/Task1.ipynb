```python
# First Question
import numpy as np
import matplotlib.pyplot as plt

# Step 1: True function (smooth sine curve)
x = np.linspace(0, 1, 100)            # 100 evenly spaced points
y = np.sin(2 * np.pi * x)         # True underlying sine function

# Step 2: Observed data (sample points + noise)
x_obs = np.linspace(0, 1, 10)               # 10 sample points
y_obs = np.sin(2 * np.pi * x_obs) + np.random.normal(0, 0.1, x_obs.shape)
# noise with mean=0, std=0.1

# Step 3: Plot
plt.figure(figsize=(6,4))
plt.plot(x, y, label='true underlying function')   # smooth sine curve
plt.scatter(x_obs, y_obs, color='blue', label='observed data') # noisy points
plt.title('Plot of noise corrupted data')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
# Second Question
input_list = input("Enter numbers separated by spaces: ")
input_list = list(map(int, input_list.split()))  # convert to list of integers

a = np.array(input_list)
b = np.append(a[1:], 0)
c = (b-a) [:-1]

if np.all(c >= 0):
  print(f'{input_list} is Monotonic(increasing)')
elif np.all(c <= 0):
  print(f'{input_list} is Monotonic(decreasing)')
else:
  print(f'{input_list} is Not monotonic')
```
